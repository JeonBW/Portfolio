# SVM (Support Vector Machine)

```python
from sklearn.datasets import make_blobs

X, y = make_blobs(centers=4 , random_state=8)
y = y % 2
mglearn.discrete_scatter(X[:,0], X[:,1],y)
```

```python
from sklearn.svm import LinearSVC
# LinearSVC는 SVM 중 하나로 분류를 위한 선형모델
```

```python
linear_svm = LinearSVC().fit(X, y)

mglearn.discrete_scatter(X[:,0], X[:,1],y)
mglearn.plots.plot_2d_separator(linear_svm, X)
# 그나마 가장 최선인 선 
```

```python
X_new = np.hstack([X, X[:, 1:] ** 2])

from mpl_toolkits.mplot3d import Axes3D, axes3d
figure = plt.figure()
# 3차원 그래프
ax = Axes3D(figure, elev=-152, azim=-26)

mask = y == 0
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b', cmap=mglearn.cm2, s=60,  edgecolor='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^', 
           cmap=mglearn.cm2, s=60, edgecolor='k')
ax.set_xlabel("특성0")
ax.set_ylabel("특성1")
ax.set_zlabel("특성1 ** 2")
```

```python
linear_svm_3d = LinearSVC().fit(X_new, y)
coef, intercept = linear_svm_3d.coef_.ravel(), linear_svm_3d.intercept_
# 선형 결정 경계 그리기
figure = plt.figure()
ax = Axes3D(figure, elev=-152, azim=-26)
xx = np.linspace(X_new[:, 0].min() - 2, X_new[:, 0].max() + 2, 50)
yy = np.linspace(X_new[:, 1].min() - 2, X_new[:, 1].max() + 2, 50)

XX, YY = np.meshgrid(xx, yy)
ZZ = (coef[0] * XX + coef[1] * YY + intercept) / -coef[2]
ax.plot_surface(XX, YY, ZZ, rstride=8, cstride=8, alpha=0.3)
ax.scatter(X_new[mask, 0], X_new[mask, 1], X_new[mask, 2], c='b', cmap=mglearn.cm2, s=60,  edgecolor='k')
ax.scatter(X_new[~mask, 0], X_new[~mask, 1], X_new[~mask, 2], c='r', marker='^', 
           cmap=mglearn.cm2, s=60, edgecolor='k')
ax.set_xlabel("특성0")
ax.set_ylabel("특성1")
ax.set_zlabel("특성1 ** 2")
```

```python
ZZ = YY ** 2
dec = linear_svm_3d.decision_function(np.c_[XX.ravel(), YY.ravel(), ZZ.ravel()])
plt.contourf(XX, YY, dec.reshape(XX.shape), levels=[dec.min(), 0, dec.max()], 
            cmap=mglearn.cm2, alpha=0.5)
mglearn.discrete_scatter(X[:, 0], X[:, 1], y)
plt.xlabel("특성 0")
plt.ylabel("특성 1")
```

### 파라미터 : cost, gamma 
    - cost : 선과 데이터 사이의 간격(마진의 폭)을 좁게 할것인지 크게 할 것인지 
    - 비용을 크게하면 마진이 작아진다.
    - 코스트를 너무 작게해도 너무 크게해도 안된다.(가급적이면 비용은 적게들게 하는 편이 좋다)
    - 코스트가 작으면 모델이 단순해져 과적합이 일어나지 않는다.
    - 코스트가 많으면 모델이 복잡해지고 과적합이 일어난다. 
    
    - 감마 : 하나의 데이터가 영향을 끼치는 범위를 지정 하는 것
    - 감마가 적을때는 영향이 적다 -> 범위가 넓어진다.
    - 감마가 크면 자기만의 영역을 다 따로따로 가진다 -> 범위가 좁아진다.
    - 적정한 수준에서 감마를 조절해야한다. 
    
    - 그리드 서치를 이용하여 최적의 파라미터를 찾는다.

---
### 파라미터 : cost, gamma 
+ cost
    - decision boundary와 마진의 간격을 어느정도로 할지 결정
    - cost가 작으면 마진은 넓어진다. 
    

+ gamma 
    - 훈련데이터 하나 당 열향을 끼치는 범위를 조절하는 변수 
    - gamma가 적을 경우 영향을 끼치는 범위가 넓어진다.

```python
from sklearn.datasets import load_breast_cancer

cancer = load_breast_cancer()
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, random_state=0) 
```

```python
svm = SVC()  #기본값 C=1 / gamma = auto(1/특성의 갯수(X_train.shape[1]))
svm.fit(X_train, y_train)
```

```python
print("훈련 결과:", svm.score(X_train, y_train))
print("테스트 결과:", svm.score(X_test, y_test))
```

```python
# 다른 알고리즘에 비해 전처리에 엄청 예민하다. 
# 전처리(표준화, 정규화, 스케일링)가 잘 못되있거나 안되있으면 안좋은 결과가 나온다. 

##전처리로 스케일 조정 
# 특성별 최소값
min_on_training = X_train.min(axis=0)
print(min_on_training)

print("----------------------------------------------------------------------------")

# 특성별 최대값
max_on_training = X_train.max(axis=0)
print(max_on_training)

print("----------------------------------------------------------------------------")

# 특성별 범위(최대값 - 최소값) 
#range_on_training = max_on_training - min_on_training
range_on_training = (X_train - min_on_training).max(axis=0)
print(range_on_training)


X_train_scaled = (X_train -  min_on_training) / range_on_training
X_test_scaled = (X_test -  min_on_training) / range_on_training # range_on_training은 훈련 데이터로 하는 것이 정설
```

```python
svm2 = SVC().fit(X_train_scaled, y_train)
```

```python
print("훈련 결과:", svm2.score(X_train_scaled, y_train))
print("테스트 결과:", svm2.score(X_test_scaled, y_test))
```

```python
# Cost로 성능 튜닝 
svm50 = SVC(C=50).fit(X_train_scaled, y_train)
```

```python
print("훈련 결과:", svm50.score(X_train_scaled, y_train))
print("테스트 결과:", svm50.score(X_test_scaled, y_test))
```

```python
print(" 최적의 하이퍼 파라미터\n", grid_cv.best_params_)
print(" 최고 예측 정확도 :", grid_cv.best_score_)
```

#### 이미지 내부의 문자 인식 (실전개발 입문 180p)

```python
import urllib.request as req
import gzip, os, os.path
savepath = "./data/mnist"
baseurl = "http://yann.lecun.com/exdb/mnist"
files = ["train-images-idx3-ubyte.gz","train-labels-idx1-ubyte.gz","t10k-images-idx3-ubyte.gz","t10k-labels-idx1-ubyte.gz"]
# 다운로드
if not os.path.exists(savepath): os.mkdir(savepath)
for f in files:
    url = baseurl + "/" + f
    loc = savepath + "/" + f
    print("download:", url)
    if not os.path.exists(loc):
        req.urlretrieve(url, loc)
# GZip 압축 해제
for f in files:
    gz_file = savepath + "/" + f
    raw_file = savepath + "/" + f.replace(".gz", "")
    print("gzip:", f)
    with gzip.open(gz_file, "rb") as fp:
        body = fp.read()
        with open(raw_file, "wb") as w:
            w.write(body)
print("ok")
```

```python
import struct
def to_csv(name, maxdata):
    # 레이블 파일과 이미지 파일 열기
    lbl_f = open("./data/mnist/"+name+"-labels-idx1-ubyte", "rb")
    img_f = open("./data/mnist/"+name+"-images-idx3-ubyte", "rb")
    
    csv_f = open("./data/mnist/"+name+".csv", "w", encoding="utf-8")
    # 헤더 정보 읽기 --- (※1)
    mag, lbl_count = struct.unpack(">II", lbl_f.read(8))
    mag, img_count = struct.unpack(">II", img_f.read(8))
    rows, cols = struct.unpack(">II", img_f.read(8))
    pixels = rows * cols
    # 이미지 데이터를 읽고 CSV로 저장하기 --- (※2)
    res = []
    for idx in range(lbl_count):
        if idx > maxdata: break
        label = struct.unpack("B", lbl_f.read(1))[0]
        bdata = img_f.read(pixels)
        sdata = list(map(lambda n: str(n), bdata))
        csv_f.write(str(label)+",")
        csv_f.write(",".join(sdata)+"\r\n")
        # 잘 저장됐는지 이미지 파일로 저장해서 테스트하기 -- (※3)
        if idx < 10:
            s = "P2 28 28 255\n"
            s += " ".join(sdata)
            iname = "./data/mnist/{0}-{1}-{2}.pgm".format(name,idx,label)
            with open(iname, "w", encoding="utf-8") as f:
                f.write(s)
    csv_f.close()
    lbl_f.close()
    img_f.close()
# 결과를 파일로 출력하기 --- (※4)
to_csv("train", 10000)
to_csv("t10k", 500)
```

```python
"""
포터블 비트맵(Potable Bitmaps) (가볍게 주고 받을 수 있는 비트맵 파일)
1. pgm : 한 화소(픽셀)를 1byte로 표현 (흑백 이미지)
    - P2, P5 
2. pbm : 한 화소를 1bit으로 표현 (흑백 이미지, pgm보다 훨신 더 작은)
    - P1, P4
3. ppm : 한 화소를 3byte로 표현 (컬러 이미지, RGB 각가 1바이트씩 )
    - P3, P6
"""
```

```python
from sklearn import model_selection, svm, metrics

def load_csv(fname):
    labels = []
    images = []
    
    with open(fname, "r") as f:
        for line in f:
            cols = line.split(",")
            if len(cols) < 2:
                continue
            
            labels.append(int(cols.pop(0)))
            
            vals = list(map(lambda n:int(n)/256, cols))
            images.append(vals)
    return {"labels":labels, "images":images}

data = load_csv("./data/mnist/train.csv")
test = load_csv("./data/mnist/t10k.csv")
```
